## 合并数据

- pd.concat()
- pd.merge()

### pd.concat()

重要参数说明：

- objs：series，dataframe或者是panel构成的序列lsit 
- axis：需要合并链接的轴，0是行，1是列 
- join：连接的方式 inner，或者oute
- ignore_index：忽略需要连接的frame本身的index，当原本的index没有特别意义的时候可以使用，默认为False

#### 合并行

````python
concat_1_10 = senior_score.iloc[0:10]
concat_11_20 = senior_score.iloc[10:20]
concat_21_30 = senior_score.iloc[20:30]
````

````python
pd.concat([concat_1_10, concat_11_20, concat_21_30], axis=0)
````

#### 合并列

````python
concat_math = senior_score.iloc[0:10, 3]
concat_physical = senior_score.iloc[0:10, 5]
concat_chemistry = senior_score.iloc[0:10, 6]
````

````python
pd.concat([concat_math, concat_physical, concat_chemistry], axis=1)
````

### pd.merge()

重要参数说明：

- left与right：两个不同的DataFrame
- how：指的是合并(连接)的方式有inner(内连接),left(左外连接),right(右外连接),outer(全外连接);默认为inner
- on：指的是用于连接的列索引名称。必须存在右右两个DataFrame对象中，如果没有指定且其他参数也未指定则以两个DataFrame的列名交集做为连接键

````python
merge_kh_chinese = senior_score.iloc[0:10, [1, 2]]
merge_kh_math = senior_score.iloc[0:10, [1, 3]]
merge_kh_english = senior_score.iloc[0:10, [1, 4]]
````

````python
pd.merge(merge_kh_chinese, merge_kh_math)
````

### 复合条形图

````python
x = '语文 数学 英语'.split(' ')
y_pink = np.random.randint(0, 150, 3)
y_skyblue = np.random.randint(0, 150, 3)
x_len = np.arange(len(y_pink))
width = 0.35

plt.figure(figsize=(12, 8))
plt.bar(x_len, y_pink, width, color='pink')
plt.bar(x_len+width, y_skyblue, width, color='skyblue')
plt.ylim(0, 150)
plt.xticks(x_len+width, x)
````

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


courses = '语文 数学 英语'.split(' ')
pink_score = np.random.randint(0, 150, 3)
skyblue_score = np.random.randint(0, 150, 3)

x = np.arange(len(courses))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots(figsize=(16, 10))
rects1 = ax.bar(x - width/2, pink_score, width, label='pink', color='pink')
rects2 = ax.bar(x + width/2, skyblue_score, width, label='skyblue', color='skyblue')
# rects1 = ax.bar(x, pink_score, width, label='pink', color='pink')
# rects2 = ax.bar(x + width, skyblue_score, width, label='skyblue', color='skyblue')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylim(0, 150)
ax.set_ylabel('Scores')
ax.set_title('成绩表')
ax.set_xticks(x)
ax.set_xticklabels(courses)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
````

## 直方图

### 频数直方图

````python
salays = np.random.RandomState(42).randint(2000, 10000, 200)
hist = plt.hist(salays, bins=8, edgecolor='k', color='w')
plt.xticks(hist[1])
plt.yticks(hist[0])
plt.axhline(y=hist[0][0], xmin=0, xmax=0.05, color='red', linestyle=':')

start = 0.05
for i in range(len(hist[0])):
    xmax = start + (0.9/len(hist[0]))*i
    plt.axhline(y=hist[0][i], xmin=0, xmax=xmax, color='red', linestyle=':')
plt.show()
````
